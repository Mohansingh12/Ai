```python
from sklearn.preprocessing import  OneHotEncoder
```

```python
he=OneHotEncoder(handle_unknown='ignore',sparse_output=False).set_output(transform='pandas')
```

```python
hedf=he.fit_transform(df[['person_gender','person_education','person_home_ownership','loan_intent','previous_loan_defaults_on_file']])
```

```python
hedf
```

```python
df=pd.concat([hedf,df],axis=1).drop(columns=['person_gender','person_education','person_home_ownership','loan_intent','previous_loan_defaults_on_file'],axis=1)
```
