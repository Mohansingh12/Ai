```python
from sklearn.preprocessing import  OneHotEncoder
```

```python
he=OneHotEncoder(handle_unknown='ignore',sparse_output=False).set_output(transform='pandas')
```

```python
hedf=he.fit_transform(df[['person_gender','person_education','person_home_ownership','loan_intent','previous_loan_defaults_on_file']])
```

```python
hedf
```

```python
df=pd.concat([hedf,df],axis=1).drop(columns=['person_gender','person_education','person_home_ownership','loan_intent','previous_loan_defaults_on_file'],axis=1)
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['loan_status'],axis=1)
```

```python
y=df['loan_status']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X_scaled,y,test_size=.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
