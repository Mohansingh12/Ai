```python
from sklearn.preprocessing import OneHotEncoder
```

```python
onehot = OneHotEncoder(handle_unknown='ignore',sparse_output=False).set_output(transform='pandas')
```

```python
ndp=onehot.fit_transform(dp[['gender','part_time_job','diet_quality','parental_education_level','internet_quality','extracurricular_participation']])
```

```python
dp.info()
```

```python
dp=pd.concat([dp,ndp],axis=1).drop(columns=['student_id','gender','part_time_job','diet_quality','parental_education_level','internet_quality','extracurricular_participation'])
```

```python
dp.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = dp.drop('exam_score', axis=1)
y = dp['exam_score']
X.shape, y.shape
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
lr_model.score(X_test, y_test)
```
