```python
import pandas as pd
```

```python
df=pd.read_csv('Titanic-Dataset.csv')
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oh=OneHotEncoder(handle_unknown='ignore',sparse_output=False).set_output(transform='pandas')
```

```python
ohdf=oh.fit_transform(df[['Sex','Embarked']])
```

```python
df=pd.concat([df,ohdf],axis=1).drop(df[['Sex','Embarked']],axis=1)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()  
```

```python
mean_val = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_val)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['Survived','Name','Ticket','PassengerId','Cabin'],axis=1)
```

```python
y=df['Survived']
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(class_weight='balanced',random_state=42)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
